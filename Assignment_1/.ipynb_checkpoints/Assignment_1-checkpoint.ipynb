### Data preparation
- Data preparation contains several steps:
- Downloading the dataset
- Re-format the dataset to fit ML model
- Vector transformation, feature and lable extraction. 

```python
import urllib.request
from scipy.io.arff import loadarff
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder

# Download the dataset from website
data_urls = ["""https://cl.lingfil.uu.se/~frewa417/english_past_tense.arff""", """https://cl.lingfil.uu.se/~frewa417/german_plural.arff"""]

filenames = [url.split("/")[-1] for url in data_urls]

for url, fn in zip(data_urls, filenames):
  urllib.request.urlretrieve(url, fn)

# Re-format the dataset to fit ML model
loaded_data_files = [loadarff(fn) for fn in filenames]
D = dict()
for data in loaded_data_files:
  data_points = data[0]
  field_names = data[1].names()
  assert field_names[0] == 'frequency'
  assert field_names[-1] == 'class'
  X = list()
  y = list()
  for point in data_points:
    v = [field_names[i]+"_"+point[i].decode("utf-8") for i in range(1, len(point)-1)]
    X.extend([v]*int(point[0]))
    assert len(v) == len(X[0])
    u = [point[-1].decode("utf-8")]
    y.extend([u]*int(point[0]))
    assert len(u) == len(y[0])
  assert len(X) == np.sum(np.asarray([point[0] for point in data_points]))
  X_orig = np.asarray(X)
  y_orig = np.asarray(y).ravel()
  D[data[1].name] = tuple([X_orig, y_orig])

# Perform verctor transformation for the dataset
feature_encoder = OneHotEncoder()
label_encoder = LabelEncoder()

X_past = feature_encoder.fit_transform(D['past-tense'][0])
names_past = feature_encoder.get_feature_names()
y_past = label_encoder.fit_transform(D['past-tense'][1])

X_plural = feature_encoder.fit_transform(D['plural'][0])
names_plural = feature_encoder.get_feature_names()
y_plural = label_encoder.fit_transform(D['plural'][1])
```

## Decision Tree

#### How accurate are the decision tree classifiers for the two data sets? Look at overall accuracy as well as precision and recall for specific classes.


The overall accuracy is caculated as below:

The max_depth is not set which means to expand the tree fully. This will result in a accuracy of 99% of accuracy. I think the reason is that amost all the data in the test dataset has been seen in the training data.  

The parameter splitter = 'random' and ccp_alpha with different has been use for trial, as the character of the dataset(no overfitting) the use of ccp_alpha will decrease the accuracy.  

The ccp_alpha parameter is tuned as a threashold to the process of this tree pruning process, which means, the bigger the ccp_alpha parameter, more tree brances will be pruned.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold
classifier_past = DecisionTreeClassifier(splitter='random')
classifier_past.fit(X_past, y_past)

for train, test in KFold(n_splits=5, shuffle=True).split(X_past):
  X_past_train = X_past[train]
  X_past_test = X_past[test]
  y_past_train = y_past[train]
  y_past_test = y_past[test]
  classifier_past.fit(X_past_train, y_past_train)
  print("Accuracy: %.1f%%" %(100*classifier_past.score(X_past_test, y_past_test)))
```

#### How does training error relate to test error?

Trainning error is the error that occurs when trainning the machine learning model using training dataset. It means the classification error(for classification model) of a model on the same data the model was trained. 
Test error is the error occurs when using the test dataset. 
For example, if a model's training error is low, but the test error is high, that means the model is likely overfitting. 

#### Can you make sense of the rules implicit in the trees?

As describe in the dataset for English past tense, p1-p3: antepenultimate, p4-p6: penultimate, p7-p9: ultimate. And in the tree described below, gini is a parameter indicating possible tree branch of a node. if gini is next to zero, the possibility of sub catergories(tree branch) for a node is low, vice versa if gini is next to 1.  
We can see that, there is high chances the classifer will add class 19 or class 41, I think the reason is that, in English past tense, high chances of words will have past tense of "adding d" and "adding ed" in the end of the words. 

For the tree, due to the huge tree, we just take one example. At the top node, if x6_p7_b <= 0.508(the p7 attribute is a and its vector value is less than 0.508), the word could be a class 19. But since gini 0.59, there is high possibility of subcategories(tree branching) under this node.

## K-Nearest Neighbor

```python
def sample_data(X, y, degree):
    I = np.random.uniform(0, 1, size=X.shape[0]) < degree
    return X[I, :], y[I]
```

```python
def get_accuracy(n_neighbors, X, y):
    knn_past = KNeighborsClassifier(n_neighbors=n_neighbors)
    sum_accuracy = 0
    for train, test in KFold(n_splits=5, shuffle=True).split(X):
        X_train = X[train]
        X_test = X[test]
        y_train = y[train]
        y_test = y[test]
        knn_past.fit(X_train, y_train)
        current_accuracy = 100*knn_past.score(X_test, y_test)
        sum_accuracy = sum_accuracy + current_accuracy
    return sum_accuracy/5
```

#### Can you improve accuracy by removing less informative features?

English past tense is used as feature exploring for this question. By guessing, English words's p7-p9: ultimate have more weight on effecting the their past tense, meanwhile their antepenultimate may have less impact.

```python
def dataset_remove_feature(column_number):
    tmp = D['past-tense'][0]
    tmp = np.delete(tmp, column_number, axis=1)
    return feature_encoder.fit_transform(tmp)
    
    
```

```python
def after_rm_accuracy(column_number):
    X_past_rm = dataset_remove_feature(column_number)
    sample_rm = sample_data(X_past_rm, y_past, .01)
    X_past_sample_rm = sample_rm[0]
    y_past_sample_rm = sample_rm[1]
    return get_accuracy(5, X_past_sample_rm, y_past_sample_rm)
```

## VG assignment

In order to save the running time, only German words dataset are choosen. A learning curve from sklearn is deployed in this assignment. train_sizes=np.linspace(.1, 1.0, 5)) is used to generate learning curve with different training dataset. As the result show, because of the dataset, dataset with different size perfomed well(Decision Tree over 99% and KNN over 95%). From the observation, we know that at the beginnign, with the increasing amount of dataset, the performance improved. Meanwhile, the improvement is not obvious in a later stage. 

```python
from sklearn.model_selection import learning_curve

# reference: https://scikit-learn.org/stable/auto_examples/model_selection/plot_learning_curve.html#sphx-glr-auto-examples-model-selection-plot-learning-curve-py
def plot_learning_curve(estimator, title, X, y, ylim=None, cv=None,
                        n_jobs=None, train_sizes=np.linspace(.1, 1.0, 5)):
    


    plt.title(title)
    if ylim is not None:
        plt.ylim(*ylim)
    plt.xlabel("Training examples")
    plt.ylabel("Score")

    train_sizes, train_scores, test_scores, fit_times, _ = \
        learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs,
                       train_sizes=train_sizes,
                       return_times=True)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    fit_times_mean = np.mean(fit_times, axis=1)
    fit_times_std = np.std(fit_times, axis=1)

    # Plot learning curve
    plt.grid()
    plt.fill_between(train_sizes, train_scores_mean - train_scores_std,
                         train_scores_mean + train_scores_std, alpha=0.1,
                         color="r")
    plt.fill_between(train_sizes, test_scores_mean - test_scores_std,
                         test_scores_mean + test_scores_std, alpha=0.1,
                         color="g")
    plt.plot(train_sizes, train_scores_mean, 'o-', color="r",
                 label="Training score")
    plt.plot(train_sizes, test_scores_mean, 'o-', color="g",
                 label="Cross-validation score")
    plt.legend(loc="best")

    return plt
```

```python
from sklearn.model_selection import ShuffleSplit

title = "Learning Curves (KNN with German words)" 
cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
plot_learning_curve(knn_plural, title, X_plural_sample, y_plural_sample,
                    cv=cv, n_jobs=4)

plt.show()
```

```python
from sklearn.model_selection import ShuffleSplit

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)

title = "Learning Curves (Decision Tree with German words)"
plot_learning_curve(classifier_plural, title, X_plural, y_plural,
                    cv=cv, n_jobs=4)
```
